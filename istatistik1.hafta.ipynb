```python
import pandas as pd 
import seaborn as sns
import numpy as np 
from matplotlib import pyplot as plt 
from scipy import stats
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
a = ["poor","Below average","average", "above average","excellent" ]


rates = [2,3,5,9,1]
```

```python
plt.bar(a,rates)
plt.xlabel("comments")
plt.ylabel("rates")
```

```python
df = pd.DataFrame(a, columns = ["comments"])
df_1 = pd.DataFrame(a, columns = ["comments"])
```

```python
df["rates"] = rates
df_1["rates"] = rates
```

```python
df_1
df_2 = df.groupby(["comments"]).sum()
df_2
```

```python
le = preprocessing.LabelEncoder()
le.fit(df.comments)
y = le.fit_transform(df.comments)
```

```python
df["comments"] = y
df
```

 commentsleri kategorik veriden numeric verilere çeviricem bunun için sckitlearn librarysinden label encoderi kullanacagım

```python
plt.figure(figsize = (12,6))
df_2.plot(kind  = "pie" , labels = df.rates,subplots = True)
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.barh(df_1.comments,df_1.rates)
plt.xlabel("comments")
plt.ylabel("rates")
plt.show()
```

```python
dif = pd.DataFrame({"State" : ["Florida", "Texas" , "Vermont"],
                   "Urban" : [65,70,55],
                    "Rural" : [70,70,65]
                   })
```

```python
dif
```

çift  bar grafiği için columnsları liste olarak ele almamız faydalı olacaktır

```python
states = dif.State.to_list()
```

```python
urbans = dif.Urban.to_list()
```

```python
rurals = dif.Rural.to_list()
```

```python
plt.figure(figsize = (14,8))
x_axis = np.arange(len(states))
plt.bar(x_axis - 0.2 ,  urbans , 0.4 , label = "urbans"  )
plt.bar(x_axis + 0.2 , rurals ,  0.4 , label = "rurals")
plt.xticks(x_axis,states)
plt.xlabel("States")
plt.ylabel("Speed Limits")
plt.title(" the highway speed limits on interstate roads")
plt.legend()
plt.show()
```
